```python
import numpy as np
import matplotlib.pyplot as plt
import sys 
import h5py
sys.path.append("../lib")
from figparams import *
plt.rcParams.update(rc_params)

with h5py.File('../data/dataFigure2a.h5', 'r') as f:
    bins = f['bins'][:]
    counts8 = f['counts8'][:]
    counts6 = f['counts6'][:]
    counts4 = f['counts4'][:]
    
fig = plt.figure(2, figsize=rect_size)

plt.step(bins[1:], np.sum(counts4, axis=1)/11465.0, where='post', color="C0",label='$\\rho_0 = 4$', lw=1.0)
plt.step(bins[1:], np.sum(counts6, axis=1)/11465.0,where='post', color="C1",label='$\\rho_0 = 6$', lw=1.0)
plt.step(bins[1:], np.sum(counts8, axis=1)/11465.0, where='post', color="C2",label='$\\rho_0 =8$', lw=1.0)

plt.axvline(4, color='black', linestyle='--', label='$T_{\\rm obs}$')
plt.legend(loc='upper right')
plt.xlabel('Time to coalescence [yr]')
plt.ylabel("Expected \# of events")

plt.xscale('log')
plt.ylim(0, 1.4)
plt.savefig('../plots/Figure2a.pdf', bbox_inches='tight')
```
