```python
import numpy as np
import sys 
import h5py 
import matplotlib.pyplot as plt
sys.path.append("../lib")
from figparams import *
plt.rcParams.update(rc_params)
```

```python
data = h5py.File('../data/dataFigure3.h5', 'r')
```

```python
fig = plt.figure(3, figsize=rect_size)

plt.grid(axis='y', which="both")
plt.bar(np.arange(4)-0.2, data['hist4binned'][()], width=0.2, label='$\\rho_0 = 4$', zorder=3, fill=True, alpha=0.5, edgecolor="C0", linewidth=1.0)
plt.bar(np.arange(4), data['hist6binned'][()], width=0.2,  label='$\\rho_0 = 6$', zorder=3, edgecolor="C1", fill=True, alpha=0.5, linewidth=1.0)
plt.bar(np.arange(4)+0.2, data['hist8binned'][()], width=0.2,  label='$\\rho_0 = 8$', zorder=3, edgecolor="C2", fill=True, alpha=0.5, linewidth=1.0)
plt.errorbar(np.arange(4)-0.2, data['hist4binned'][()], yerr=[data['hist4binned'][()]-data['lower4'][()], data['upper4'][()]-data['hist4binned'][()]], fmt='s', zorder=4, capsize = 5, color='C0')             
plt.errorbar(np.arange(4), data['hist6binned'][()], yerr=[data['hist6binned'][()]-data['lower6'][()], data['upper6'][()]-data['hist6binned'][()]], fmt='s', zorder=4, capsize=3, color="C1", )
plt.errorbar(np.arange(4)+0.2, data['hist8binned'][()], yerr=[data['hist8binned'][()]-data['lower8'][()], data['upper8'][()]-data['hist8binned'][()]], fmt='s', zorder=4, capsize=3, color="C2", )
yloc= 6e-2
cnts = data['counts8binned'][()]
for i in range(4):
    val=np.round(100-np.count_nonzero(cnts[i,:])/cnts.shape[1]*100,1)
    plt.text(i+0.2, yloc, f"$ {val} \%$", color='k', fontsize=10, rotation=90, ha='center', va='bottom', 
             zorder=10, bbox=dict(facecolor="white", edgecolor='black', boxstyle='round,pad=0.5'))
cnts = data['counts6binned'][()]
for i in range(4):
    val=np.round(100-np.count_nonzero(cnts[i,:])/cnts.shape[1]*100,1)
    plt.text(i-0.2, yloc, f"$ {val} \%$", color='k', fontsize=10, rotation=90, ha='center', va='bottom',
             zorder=10, bbox=dict(facecolor="white", edgecolor='black', boxstyle='round,pad=0.5'))
cnts = data['counts6binned'][()]
for i in range(4):
    val=np.round(100-np.count_nonzero(cnts[i,:])/cnts.shape[1]*100,1)
    plt.text(i+0.0, yloc, f"$\, {val} \% \,$", color='k', fontsize=10, rotation=90, ha='center', va='bottom',
             zorder=10, bbox=dict(facecolor="white", edgecolor='black', boxstyle='round,pad=0.5'))
plt.legend(loc='upper center', framealpha=0.5)
plt.yscale('log')
plt.ylim(5e-2, 100)
plt.xticks(np.arange(4), ['0 to 5', '5 to 10', '10 to 15', '$>15$'])
plt.xlabel('Time to coalescence [yr]')
plt.ylabel("\# of events")
plt.savefig('../plots/Figure2b.pdf', bbox_inches='tight')
```
