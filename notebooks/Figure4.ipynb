```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import h5py
from matplotlib.legend_handler import HandlerTuple
```

```python
import sys 
sys.path.append("../lib")
from figparams import *
plt.rcParams.update(rc_params)
```

```python
data = h5py.File('../data/dataFigure4.h5', 'r')

cols = ['Redshift', 'Mass1', 'Mass2', 'Distance', 'CoalTime', 'ChirpMass',
       'InitialFrequency', 'FinalFrequency', 'InBandTime', 'Spin1', 'Spin2',
       'AzimuthalAngleOfSpin1', 'AzimuthalAngleOfSpin2', 'InitialPhase',
       'EclipticLatitude', 'EclipticLongitude', 'Inclination', 'Polarization',
       'snr_avg_inclination', 'snr_max_inclination', 'snr_LISA', 'catalogue',
       'Amplitude', 'FrequencyDerivative']
ev8_10 = pd.DataFrame(data['catalogue8_10'][:], columns=cols)
ev8_5 = pd.DataFrame(data['catalogue8_5'][:], columns=cols)
ev8_1 = pd.DataFrame(data['catalogue8_1'][:], columns=cols)
data.close()
```

```python
fig, ax = plt.subplots(2, 2, figsize=square_size, sharex='col', sharey='row')
col1 = [plt.cm.Blues(x) for x in np.linspace(0.3, 1, 5)]
col5 = [plt.cm.Greens(x) for x in np.linspace(0.3, 1, 5)]
col10 = [plt.cm.Reds(x) for x in np.linspace(0.3, 1, 5)]
kwargs1 = {'s': 50, 'alpha': 0.7, 'edgecolor': 'k', 'linewidth': 0.5, 'marker':'o'}
kwargs5 = {'s': 50, 'alpha': 0.7, 'edgecolor': 'k', 'linewidth': 0.5, 'marker':'s'}
kwargs10 = {'s': 50, 'alpha': 0.7, 'edgecolor': 'k', 'linewidth': 0.5, 'marker':'^'}
xcon = 'InitialFrequency'
ycon = 'Amplitude'
catnum10 = np.unique(ev8_10['catalogue'])
for id, cat in enumerate(catnum10):
    for i, ev in ev8_10[ev8_10['catalogue']==cat].iterrows():
        ax[0,0].scatter(ev[xcon]*1000, ev[ycon], color=col10[id], **kwargs10)
catnum5 = np.unique(ev8_5['catalogue'])
for id, cat in enumerate(catnum5):
    for i, ev in ev8_5[ev8_5['catalogue']==cat].iterrows():
        ax[0,0].scatter(ev[xcon]*1000, ev[ycon], color=col5[id], **kwargs5)
for id, (i, ev) in enumerate(ev8_1.iterrows()):
    ax[0,0].scatter(ev[xcon]*1000, ev[ycon], color=col1[id], **kwargs1)
ax[0,1].axis('off')
xcon = 'InitialFrequency'
ycon = 'FrequencyDerivative'
catnum10 = np.unique(ev8_10['catalogue'])
for id, cat in enumerate(catnum10):
    for i, ev in ev8_10[ev8_10['catalogue']==cat].iterrows():
        ax[1,0].scatter(ev[xcon]*1000, ev[ycon], color=col10[id], **kwargs10)
catnum5 = np.unique(ev8_5['catalogue'])
for id, cat in enumerate(catnum5):
    for i, ev in ev8_5[ev8_5['catalogue']==cat].iterrows():
        ax[1,0].scatter(ev[xcon]*1000, ev[ycon], color=col5[id], **kwargs5)
for id, (i, ev) in enumerate(ev8_1.iterrows()):
    ax[1,0].scatter(ev[xcon]*1000, ev[ycon], color=col1[id], **kwargs1)
xcon = 'Amplitude'
ycon = 'FrequencyDerivative'
catnum10 = np.unique(ev8_10['catalogue'])
for id, cat in enumerate(catnum10):
    for i, ev in ev8_10[ev8_10['catalogue']==cat].iterrows():
        ax[1,1].scatter(ev[xcon], ev[ycon], color=col10[id], **kwargs10)
catnum5 = np.unique(ev8_5['catalogue'])
for id, cat in enumerate(catnum5):
    for i, ev in ev8_5[ev8_5['catalogue']==cat].iterrows():
        ax[1,1].scatter(ev[xcon], ev[ycon], color=col5[id], **kwargs5)
for id, (i, ev) in enumerate(ev8_1.iterrows()):
    ax[1,1].scatter(ev[xcon], ev[ycon], color=col1[id], **kwargs1)
ax[0,0].set_ylabel('$\log_{10}\mathcal{A}$')
ax[1,0].set_ylabel('$\log_{10}(\dot{f}_0/{\\rm Hz^2})$')
ax[1,0].set_xlabel('$f_0$ [mHz]')
ax[1,1].set_xlabel('$\log_{10}\mathcal{A}$')
ax[0,0].set_ylim(-23.6, -22.4)
plt.subplots_adjust(wspace=0.05, hspace=0.05)
kwdict = dict(marker='o', color='w', markeredgecolor='k', alpha=0.7, markersize=6)
legend_entry1 = (plt.Line2D([0], [0], markerfacecolor=col1[0], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col1[1], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col1[2], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col1[3], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col1[4], **kwdict))
kwdict = dict(marker='s', color='w', markeredgecolor='k', alpha=0.7, markersize=6)
legend_entry2 = (plt.Line2D([0], [0], markerfacecolor=col5[0], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col5[1], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col5[2], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col5[3], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col5[4], **kwdict))
kwdict = dict(marker='^', color='w', markeredgecolor='k', alpha=0.7, markersize=6)
legend_entry3 = (plt.Line2D([0], [0], markerfacecolor=col10[0], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col10[1], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col10[2], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col10[3], **kwdict),
                 plt.Line2D([0], [0], markerfacecolor=col10[4], **kwdict))
ax[0,1].legend(handles=[legend_entry1, legend_entry2, legend_entry3], 
               labels=['1 event', '5 events', '10 events'],
               loc='upper right', numpoints=1, handler_map={tuple: HandlerTuple(ndivide=None)})
fig.align_ylabels(ax[:,0])
plt.savefig('../plots/Figure4.pdf', bbox_inches='tight')
```
